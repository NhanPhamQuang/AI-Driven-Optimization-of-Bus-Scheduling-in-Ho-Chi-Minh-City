```python
import requests
import json

base_url = "https://api.xebuyt.net/businfo/getvarsbyroute/{}_1"

have_data = []
no_data = []

for route_id in range(1, 201):
    print("=" * 80)
    print(f"üîπ ƒêang request RouteId {route_id}")

    url = base_url.format(route_id)
    try:
        response = requests.get(url, timeout=8)
        print(f"HTTP {response.status_code}")

        if response.status_code == 200 and response.text.strip() not in ["", "null", "[]", "{}"]:
            try:
                # Parse JSON ƒë·ªÉ t·ª± decode Unicode
                data = response.json()
                print(json.dumps(data, indent=2, ensure_ascii=False))
                have_data.append(route_id)
            except json.JSONDecodeError:
                print("(Kh√¥ng ph·∫£i JSON h·ª£p l·ªá)")
                print(response.text)
                no_data.append(route_id)
        else:
            print("(R·ªóng ho·∫∑c kh√¥ng c√≥ d·ªØ li·ªáu)")
            no_data.append(route_id)

    except requests.exceptions.RequestException as e:
        print(f"‚ö†Ô∏è L·ªói khi request RouteId {route_id}: {e}")
        no_data.append(route_id)

print("\n" + "=" * 80)
print("‚úÖ T·ªîNG K·∫æT")
print(f"C√≥ d·ªØ li·ªáu ({len(have_data)}): {have_data}")
print(f"Kh√¥ng c√≥ d·ªØ li·ªáu ho·∫∑c l·ªói ({len(no_data)}): {no_data}")
```

```python
import requests
import json
from tabulate import tabulate

def safe_get_json(url):
    """Request v√† parse JSON an to√†n, tr·∫£ [] n·∫øu c√≥ l·ªói."""
    try:
        res = requests.get(url, timeout=8)
        if res.status_code != 200:
            print(f"   ‚ö†Ô∏è HTTP {res.status_code} t·∫°i {url}")
            return []
        text = res.text.strip()
        if text in ["", "null", "[]", "{}"]:
            return []
        try:
            return res.json()
        except json.JSONDecodeError:
            print(f"   ‚ö†Ô∏è JSONDecodeError t·∫°i {url}")
            return []
    except requests.exceptions.RequestException as e:
        print(f"   ‚ö†Ô∏è RequestException t·∫°i {url}: {e}")
        return []

def get_route_variants(route_id, direction):
    url = f"https://api.xebuyt.net/businfo/getvarsbyroute/{route_id}_{direction}"
    return safe_get_json(url)

def get_stops_by_variant(route_id, direction, route_var_id):
    url = f"https://api.xebuyt.net/businfo/getstopsbyvar/{route_id}_{direction}/{route_var_id}"
    return safe_get_json(url)

def get_path_by_variant(route_id, direction, route_var_id):
    url = f"https://api.xebuyt.net/businfo/getpathbyvar/{route_id}_{direction}/{route_var_id}"
    return safe_get_json(url)

def crawl_bus_data(start_id=1, end_id=200):
    results = []
    for route_id in range(start_id, end_id + 1):
        print("=" * 80)
        print(f"üöå ƒêang x·ª≠ l√Ω RouteId {route_id}")

        for direction in [1, 2]:
            direction_name = "L∆∞·ª£t ƒëi" if direction == 1 else "L∆∞·ª£t v·ªÅ"
            print(f"üîπ {direction_name} ({direction})")

            variants = get_route_variants(route_id, direction)
            if not variants:
                print("   ‚Üí Kh√¥ng c√≥ d·ªØ li·ªáu bi·∫øn th·ªÉ.")
                continue

            for v in variants:
                route_var_id = v.get("RouteVarId")
                route_var_name = v.get("RouteVarName", "")
                start_stop = v.get("StartStop", "")
                end_stop = v.get("EndStop", "")

                stops = get_stops_by_variant(route_id, direction, route_var_id)
                path = get_path_by_variant(route_id, direction, route_var_id)

                results.append([
                    route_id,
                    direction_name,
                    route_var_id,
                    route_var_name,
                    start_stop,
                    end_stop,
                    len(stops),
                    len(path)
                ])

                print(f"   ‚Üí {route_var_name}: {len(stops)} stops, {len(path)} path points")

    return results


if __name__ == "__main__":
    all_results = crawl_bus_data(1, 200)

    print("\n" + "=" * 80)
    print("üìä T·ªîNG K·∫æT K·∫æT QU·∫¢\n")
    headers = ["RouteId", "H∆∞·ªõng", "RouteVarId", "RouteVarName", "B·∫øn ƒë·∫ßu", "B·∫øn cu·ªëi", "S·ªë tr·∫°m", "S·ªë ƒëi·ªÉm path"]
    print(tabulate(all_results, headers=headers, tablefmt="grid"))

```
